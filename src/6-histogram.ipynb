### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Data for histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 15))

# Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(categories, values, color=['blue', 'orange', 'green', 'red'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'orange', 'green', 'red'])
axs[1, 1].set_title('Pie Chart')

# Histogram
axs[2, 0].hist(data, bins=5, color='cyan', edgecolor='black')
axs[2, 0].set_title('Histogram')
axs[2, 0].set_xlabel('Values')
axs[2, 0].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
